# IMPORT REQUIRED MODULES

```python
#import required modules
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
#load data
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head(10)
```

```python
#arrange data
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

# KNN REGRESSOR

```python
#train data using knn regressor with k=5
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
knn = KNeighborsRegressor(n_neighbors = 6)
knn.fit(X_train, y_train)
#check score which is accuracy. It shows approximately 90%
knn.score(X_test, y_test)
```

```python
#predict values of test data
y_pred = knn.predict(X_test)
```

# COMPARISON OF TEST DATA AND PREDICTED DATA

```python
#comparison table to predicted score and actual score of test data
dl = pd.DataFrame({'No. of Hours':X_test.flatten(), 'Predicted Score': y_pred,'Actual Score': y_test,}) 
dl
```

# PREDICTION MODEL

```python
#take input as hours of study and predict the score of student
x=input("Enter Hours of Study:")
print("The predicted score for {} of study is {}".format(x,knn.predict(x)))
```

# GRAPH

```python
#plot a graph
T = np.linspace(0, 10, 10)[:, np.newaxis]
y_ = knn.fit(X_train, y_train).predict(T)
plt.scatter(X_train, y_train, color='darkorange', label='data')
plt.plot(T, y_, color='navy', label='prediction')

plt.axis('tight')
plt.legend()
plt.title("KNeighborsRegressor with k = 5")


plt.tight_layout()
plt.show()
```
